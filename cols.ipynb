```python
import pandas as pd

data = pd.read_csv("resources/air.csv")
data.head()
```

```python
data = data.drop(["PM10", "Air Quality"], axis=1)
data.head()
```

```python
features = data.columns
features
```

```python
# max and min values
for feature in features:
    min = data[feature].min()
    max = data[feature].max()
    print(
        f"{feature}: min: {min}, max: {max}"
    )
```
